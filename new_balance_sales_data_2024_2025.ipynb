```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns




sales_by_store_2024 = pd.DataFrame({
    'code' : ["103131", "103132", "103133", "103135",  "103136", "103137", "103138", "103139", "1031310"],
    'store' : ["MEMPHIS", "COOL SPRINGS", "NEW ORLEANS" , "BIRMINGHAM", "ATLANTA II NP", "GREENVILLE", "ATLANTA III SANDY SPRINGS", "ATLANTA IV MOG" , "ATLANTA V TOCO"],
    'sales' : [1479528.92, 1190563.45, 2285021.86, 1473751.26, 1052345.72, 1314254.18, 1057253.34, 1213819.24, 1756087.48],
    'UPT' : [1.26, 1.31, 1.20, 1.27, 1.31, 1.19, 1.22, 1.30, 1.22],
    'DPT' : [129.36, 133.52, 130.38, 133.52, 129.47, 121.78, 121.30, 130.88, 131.10]


    })

sales_by_store_2025 = pd.DataFrame({
    'code' : ["103131", "103132", "103133",  "103135",  "103136", "103137", "103138", "103139", "1031310"],
    'store' : ["MEMPHIS", "COOL SPRINGS", "NEW ORLEANS" ,  "BIRMINGHAM", "ATLANTA II NP", "GREENVILLE", "ATLANTA III SANDY SPRINGS", "ATLANTA IV MOG" , "ATLANTA V TOCO"],
    'sales' : [1620487.68, 1065340.10, 1993025.02, 1451858.03, 918023.37, 1285013.71, 1049219.42, 1148985.27, 1547186.87],
    'UPT' : [1.25, 1.29, 1.20,  1.28, 1.24, 1.19, 1.18, 1.29, 1.23],
    'DPT' : [132.47, 136.04, 127.68, 139.07, 123.42, 124.30, 123.70, 136.33, 131.62]


})
# Check lenght of columns
for key, values in sales_by_store_2024.items():
    print(f"{key}: {len(values)} values")
for key, values in sales_by_store_2025.items():
    print(f"{key}:{len(values)}values")
#Get basic info and data type
print("n/Info: ")
sales_by_store_2024.info()
print("n/Info: ")
sales_by_store_2025.info()


#View Data Set
print(sales_by_store_2024.head(5))
print(sales_by_store_2025.head(5))
print(sales_by_store_2024)
print(sales_by_store_2025)


# Transform  store code to make easier to manipulate

code_conversion = {
    "103131" : "1",
    "103132" : "2",
    "103133" : '3',
    "103135" : "5",
    "103136" : "6",
    "103137" : "7",
    "103138" : "8",
    "103139" : "9",
    "1031310": "10",

}
sales_by_store_2024 = sales_by_store_2024.replace({"code": code_conversion})
sales_by_store_2025 = sales_by_store_2025.replace({"code": code_conversion})



# Describe and getting mean , median, and mode on sales

print("\nDescribe :")
print(sales_by_store_2024.describe().T.plot(kind="bar"))
sales_by_store_2025.describe()

# Order by sales
sales_by_store_2024 = sales_by_store_2024.sort_values(by = 'sales',
                                                      ascending = False)
sales_by_store_2025 = sales_by_store_2025.sort_values(by = 'sales',
                                                     ascending = False)


# Visualize sales 2024
plt.figure(figsize = (18, 5))
plt.bar(sales_by_store_2024['code'], sales_by_store_2024['sales'])
plt.title(' store sales 2024')
plt.xlabel('code')
plt.ylabel('Sales (millions)')
plt.tight_layout()
plt.show()
# Visualize sales 2025
plt.figure(figsize = (18, 5))
plt.bar(sales_by_store_2025['store'], sales_by_store_2025['sales'])
plt.title(' store sales 2025')
plt.xlabel('Store')
plt.ylabel('Sales (millions)')
plt.tight_layout()
plt.show()


# Order by UPT
sales_by_store_2024 = sales_by_store_2024.sort_values(by = 'UPT',
                                                      ascending = False)
sales_by_store_2025 = sales_by_store_2025.sort_values(by = 'UPT',
                                                     ascending = False)


# Visualize UPT 2024
plt.figure(figsize = (18, 5))
plt.bar(sales_by_store_2024['store'], sales_by_store_2024['UPT'])
plt.title('Store UPT(Units Per Transaction) 2024')
plt.xlabel('Store')
plt.ylabel('Units Per Trasaction')
plt.show()
# Visualize UPT 2025
plt.figure(figsize = (18, 4))
plt.bar(sales_by_store_2025['store'], sales_by_store_2025['UPT'])
plt.title('Store UPT(Units Per Transaction) 2025')
plt.xlabel('Store')
plt.ylabel('Units Per Trasaction')
plt.show()

# Order by DPT
sales_by_store_2024 = sales_by_store_2024.sort_values(by = 'DPT',
                                                      ascending = False)
sales_by_store_2025 = sales_by_store_2025.sort_values(by = 'DPT',
                                                     ascending = False)
# Visualize DPT 2024
plt.figure(figsize = (18, 5))
plt.bar(sales_by_store_2024['store'], sales_by_store_2024['DPT'])
plt.title('Store DPT(Dollars Per Transaction) 2024')
plt.xlabel('Store')
plt.ylabel('Dollars Per Transaction')
plt.show()

# Visualze DPT 2025
plt.figure(figsize = (18 , 5))
plt.bar(sales_by_store_2025['code'], sales_by_store_2025['DPT'])
plt.title('Store DPT(DOllars Per Transaction) 2025')
plt.xlabel('Store Code')
plt.ylabel('Dollars Per Transaction')
plt.show()

# Merge Sales Data X
sales_24 = sales_by_store_2024.rename(columns={'sales': 'sales_2024'})
sales_25 =  sales_by_store_2025.rename(columns={'sales': 'sales_2025'})

merger_sales = sales_24.merge(sales_25 ,how = "inner", on = ['code', 'store']).sort_values(by = 'sales_2024', ascending = False)



# Melt the data for seaborn
melted_data = merger_sales.melt(
    id_vars=['store'],
    value_vars=['sales_2024', 'sales_2025'],
    var_name='year',
    value_name='sales'
)

plt.figure(figsize=(18, 5))
sns.barplot(data=melted_data, x='store', y='sales', hue='year')
plt.title('Store Sales 2024 and 2025')
plt.xlabel('Store')
plt.ylabel('Sales (millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Rename UPT

upt_24 = sales_by_store_2024.rename(columns={'UPT': 'upt_24'})
upt_25 = sales_by_store_2025.rename(columns = {'UPT': 'upt_25'})

# Merge Units Per Transaction
merge_upt = upt_24.merge(upt_25 , how = 'outer' , on = ['code', 'store']).sort_values(by = 'upt_24', ascending = False)

#Visualize UPT merge side by side
merge_upt.set_index('code')[['upt_24', 'upt_25']].plot(
    kind = 'bar',
    figsize = (18,5),
    title = 'UPT 2024 And 2025'
    )
plt.ylabel('UPT')
plt.xlabel('Store Code')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
